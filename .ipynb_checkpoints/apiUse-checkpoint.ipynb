```python
def get_teams_data(years="all", teamIDs=None, divisionIDs=None):
    """
    Fetches the team data from the UFA Stats API based on provided parameters.
    
    Parameters:
    years (str): A comma-delimited list of four digit years or "all" to include all years.
    teamIDs (str): A comma-delimited list of team IDs (optional).
    divisionIDs (str): A comma-delimited list of division IDs (optional).
    
    Returns:
    dict: The parsed JSON response from the API containing team data.
    """
    # Define the API endpoint
    url = "https://www.backend.ufastats.com/api/v1/teams"
    
    # Define the parameters for the API call
    params = {
        "years": years,
        "teamIDs": teamIDs,
        "divisionIDs": divisionIDs
    }
    
    # Make the GET request to the API
    response = requests.get(url, params=params)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Return the parsed JSON response
        return response.json()
    else:
        # Return an error message if the request failed
        return {"error": f"Failed to retrieve data: {response.status_code}"}
```

```python
team_data = get_teams_data(2025)

team_IDs = []

if "error" not in team_data:
    # Loop through the teams in the "data" field
    for team in team_data["data"]:
        #print(f"Team Name: {team['fullName']}, Wins: {team['wins']}, Losses: {team['losses']}")
        team_IDs.append({'ID': team['teamID'], 'Name': team['fullName']})
else:
    # Print the error if the request failed
    print(team_data["error"])

print(team_IDs)
```

```python
def get_games(date="2024"):
    url = 'https://www.backend.ufastats.com/api/v1/games'
    
    params = {
        "date": date
    }
    
    response = requests.get(url, params=params)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Return the parsed JSON response
        return response.json()
    else:
        # Return an error message if the request failed
        return {"error": f"Failed to retrieve data: {response.status_code}"}
```

```python
def get_games_events(game_ids):
    """
    Fetches game events from the UFA Stats API for a list of game IDs.
    
    Args:
    game_ids (list): List of game IDs to fetch the events for.
    
    Returns:
    pd.DataFrame: A DataFrame containing event data with throwerX, throwerY, receiverX, and receiverY for all games.
    """
    # Define the API endpoint
    url = 'https://www.backend.ufastats.com/api/v1/gameEvents'
    
    # Initialize an empty list to store DataFrames for each game
    all_events = []
    
    # Loop through each game ID
    for gameID in game_ids:
        
        # Define the game ID as a parameter
        params = {'gameID': gameID}
        
        try:
            # Make the GET request to the API
            response = requests.get(url, params=params)
            
            # Check if the request was successful
            if response.status_code == 200:
                # Check if the response is in JSON format
                if response.headers['Content-Type'].startswith('application/json'):
                    data = response.json()  # Parse the JSON response
                    
                    # Extract relevant event data
                    home_events = data.get("data", {}).get("homeEvents", [])
                    away_events = data.get("data", {}).get("awayEvents", [])
                    
                    # Merge both home and away events
                    events = home_events + away_events

                    # If events are available, extract coordinates
                    if events:
                        df = pd.DataFrame(events)
                        # Check for the necessary columns before extracting
                        if {'throwerX', 'throwerY', 'receiverX', 'receiverY'}.issubset(df.columns):
                            df = df[['throwerX', 'throwerY', 'receiverX', 'receiverY']]
                            all_events.append(df)  # Append the DataFrame for this game
                        else:
                            print(f"Event data for game {gameID} does not contain all required fields.")
                    else:
                        print(f"No events found for game {gameID}.")
                else:
                    print(f"Response for game {gameID} is not JSON: {response.text}")
            else:
                print(f"Request for game {gameID} failed with status code {response.status_code}")
                print("Response text:", response.text)
                
        except requests.exceptions.RequestException as e:
            print(f"Error fetching data from {url} for game {gameID}: {e}")
    
    # Concatenate all DataFrames if there are any, else return an empty DataFrame
    if all_events:
        return pd.concat(all_events, ignore_index=True)
    else:
        return pd.DataFrame()
```

```python
# Example usage with a game ID
events_df = get_games_events(game_IDs)

events_df = events_df.dropna().reset_index(drop=True).copy()
# Show the DataFrame (will be empty if there was an error)
print(events_df)
```

```python
def get_games_events_2(game_ids):
    """
    Fetches game events from the UFA Stats API for a list of game IDs.
    
    Args:
    game_ids (list): List of game IDs to fetch the events for.
    
    Returns:
    tuple: DataFrames containing event data for throws, goals, drops, throwaways, callahans, and stalls.
    """
    # Define the API endpoints
    events_url = 'https://www.backend.ufastats.com/api/v1/gameEvents'
    games_url = 'https://www.backend.ufastats.com/api/v1/games'
    
    # Retrieve game details to get team IDs and extract the year from the gameID
    game_ids_str = ','.join(game_ids)
    games_params = {'gameIDs': game_ids_str}
    try:
        games_response = requests.get(games_url, params=games_params)
        games_response.raise_for_status()
        games_data = games_response.json().get("data", [])
        
        # Map each gameID to its corresponding team IDs and year
        team_id_map = {}
        for game in games_data:
            # Extract year from gameID (first 4 characters)
            year = game["gameID"][:4]

            # Map the game to its home and away team IDs and year
            team_id_map[game["gameID"]] = {
                "home": game["homeTeamID"],
                "away": game["awayTeamID"],
                "year": year
            }
    except requests.exceptions.RequestException as e:
        print(f"Error fetching game data from {games_url}: {e}")
        return None

    # Initialize lists for each event type
    pulls = []
    throws = []
    goals = []
    drops = []
    throwaways = []
    callahans = []
    stalls = []
    
    # Loop through each game ID
    for gameID in game_ids:
        # Define the game ID as a parameter
        params = {'gameID': gameID}
        
        try:
            # Make the GET request to the API for events
            response = requests.get(events_url, params=params)
            
            # Check if the request was successful
            if response.status_code == 200 and response.headers['Content-Type'].startswith('application/json'):
                data = response.json()
                
                # Get team names and year for the current game
                home_team_id = team_id_map.get(gameID, {}).get("home", "Unknown")
                away_team_id = team_id_map.get(gameID, {}).get("away", "Unknown")
                game_year = team_id_map.get(gameID, {}).get("year", "Unknown")
                
                # Extract event data
                home_events = data.get("data", {}).get("homeEvents", [])
                away_events = data.get("data", {}).get("awayEvents", [])
                
                # Replace 'home' and 'away' with actual team IDs and add year
                for event in home_events:
                    event['team'] = home_team_id
                    event['year'] = game_year
                
                for event in away_events:
                    event['team'] = away_team_id
                    event['year'] = game_year
                
                # Combine home and away events
                events = home_events + away_events
                
                # Process each event and categorize based on event type
                if events:
                    df = pd.DataFrame(events)
                    for _, row in df.iterrows():
                        if row['type'] == 18:
                            throws.append({"type": row['type'], "thrower": row['thrower'], "throwerX": row['throwerX'], "throwerY": row['throwerY'], "receiver": row['receiver'], "receiverX": row['receiverX'], "receiverY": row['receiverY'], "team": row['team'], "year": row['year'], "gameID": gameID})
                        elif row['type'] == 19:
                            goals.append({"type": row['type'], "thrower": row['thrower'], "throwerX": row['throwerX'], "throwerY": row['throwerY'], "receiver": row['receiver'], "receiverX": row['receiverX'], "receiverY": row['receiverY'], "team": row['team'], "year": row['year'], "gameID": gameID})
                        elif row['type'] == 20:
                            drops.append({"type": row['type'], "thrower": row['thrower'], "throwerX": row['throwerX'], "throwerY": row['throwerY'], "receiver": row['receiver'], "receiverX": row['receiverX'], "receiverY": row['receiverY'], "team": row['team'], "year": row['year'], "gameID": gameID})
                        elif row['type'] == 22:
                            throwaways.append({"type": row['type'], "thrower": row['thrower'], "throwerX": row['throwerX'], "throwerY": row['throwerY'], "turnoverX": row['turnoverX'], "turnoverY": row['turnoverY'], "team": row['team'], "year": row['year'], "gameID": gameID})
                        elif row['type'] == 23:
                            callahans.append({"type": row['type'], "thrower": row['thrower'], "throwerX": row['throwerX'], "throwerY": row['throwerY'], "turnoverX": row['turnoverX'], "turnoverY": row['turnoverY'], "team": row['team'], "year": row['year'], "gameID": gameID})
                        elif row['type'] == 24:
                            stalls.append({"type": row['type'], "thrower": row['thrower'], "throwerX": row['throwerX'], "team": row['team'], "year": row['year'], "gameID": gameID})
            
                
        except requests.exceptions.RequestException as e:
            print(f"Error fetching event data from {events_url} for game {gameID}: {e}")
    
    return throws, goals, drops, throwaways, callahans, stalls
```

```python
def count_event_types(event_list, target_types):
    """
    Count how many events in the event_list have a type in target_types.
    
    Args:
        event_list (list): List of event dictionaries.
        target_types (set): Set of target event types to count.
    
    Returns:
        dict: A dictionary with keys as the target event types and values as their counts.
    """
    counts = {event_type: 0 for event_type in target_types}
    for event in event_list:
        event_type = event.get("type")
        if event_type in target_types:
            counts[event_type] += 1
    return counts

def get_games_event_counts(game_ids):
    """
    Fetch game events from the UFA Stats API for the provided game IDs and
    aggregate counts for events:
        - "given": events of types 11 and 13
        - "received": events of types 20, 22, 23, and 24
    for both home and away events.
    
    Args:
        game_ids (list): List of game IDs to fetch events for.
    
    Returns:
        dict: A mapping of gameID to a dictionary with aggregated counts for home and away events.
              Example:
              {
                "gameID1": {
                    "home": {"given": X, "received": Y},
                    "away": {"given": X, "received": Y}
                },
                "gameID2": { ... }
              }
    """
    # API endpoints
    events_url = 'https://www.backend.ufastats.com/api/v1/gameEvents'
    games_url = 'https://www.backend.ufastats.com/api/v1/games'
    
    # Get game details for mapping team IDs and year info
    game_ids_str = ','.join(game_ids)
    games_params = {'gameIDs': game_ids_str}
    try:
        games_response = requests.get(games_url, params=games_params)
        games_response.raise_for_status()
        games_data = games_response.json().get("data", [])
        
        # Build a mapping from gameID to its home and away team IDs and the game year.
        team_id_map = {}
        for game in games_data:
            # Extract year from gameID (first 4 characters)
            year = game["gameID"][:4]
            team_id_map[game["gameID"]] = {
                "home": game["homeTeamID"],
                "away": game["awayTeamID"],
                "year": year
            }
    except requests.exceptions.RequestException as e:
        print(f"Error fetching game data from {games_url}: {e}")
        return None

    # Define the target event types.
    # We will later group them into "given" and "received".
    target_types = {11, 13, 20, 22, 23, 24}
    
    # Dictionary to store the aggregated counts for each gameID.
    results = {}
    
    # Loop through each gameID to fetch and process event data.
    for gameID in game_ids:
        params = {'gameID': gameID}
        try:
            response = requests.get(events_url, params=params)
            if response.status_code == 200 and response.headers.get('Content-Type', '').startswith('application/json'):
                data = response.json()
                
                # Extract home and away events
                home_events = data.get("data", {}).get("homeEvents", [])
                away_events = data.get("data", {}).get("awayEvents", [])
                
                # Optionally, add team and year info to each event (if needed elsewhere)
                home_team = team_id_map.get(gameID, {}).get("home", "Unknown")
                away_team = team_id_map.get(gameID, {}).get("away", "Unknown")
                game_year = team_id_map.get(gameID, {}).get("year", "Unknown")
                for event in home_events:
                    event['team'] = home_team
                    event['year'] = game_year
                for event in away_events:
                    event['team'] = away_team
                    event['year'] = game_year
                
                # Count the target event types in both home and away events.
                home_counts = count_event_types(home_events, target_types)
                away_counts = count_event_types(away_events, target_types)
                
                # Aggregate counts:
                # "given" are events of type 11 and 13
                # "received" are events of type 20, 22
                home_given = home_counts.get(11, 0) + home_counts.get(13, 0)
                home_received = (home_counts.get(20, 0) +
                                 home_counts.get(22, 0))
                away_given = away_counts.get(11, 0) + away_counts.get(13, 0)
                away_received = (away_counts.get(20, 0) +
                                 away_counts.get(22, 0))
                
                # Store the aggregated counts in the results dictionary.
                results[gameID] = {
                    "home": {"given": home_given, "received": home_received},
                    "away": {"given": away_given, "received": away_received}
                }
            else:
                print(f"Non-successful response for game {gameID}")
        except requests.exceptions.RequestException as e:
            print(f"Error fetching event data for game {gameID}: {e}")
    
    return results


event_counts = get_games_event_counts(game_IDs)
print(json.dumps(event_counts, indent=4))
```

```python
def get_games_events_with_blocks(game_ids):
    """
    Fetches game events from the UFA Stats API for a list of game IDs and matches 
    turnovers (throwaways [22] and drops [20]) with possession change events 
    (switch flags [13] and blocks [11]), collecting only the relevant columns 
    for each table.

    The matching is based on the invariant:
         (#throwaways + #drops) == (#switch flags + #blocks)
    The nth turnover from one team is paired with the nth possession change 
    event from the opponent's events. If that possession change is a block (type 11), 
    the turnover record is updated with block details (using its own turnoverX/turnoverY) 
    and a block record is created with a blockID equal to the turnover's assigned throwawayID.
    
    Returns:
        tuple: (throws, goals, drops, throwaways, callahans, stalls, blocks)
    """

    # Define the columns we care about for each table:
    throws_columns      = ['type', 'thrower', 'throwerX', 'throwerY', 
                           'receiver', 'receiverX', 'receiverY', 'team', 'year', 'gameID']
    throwaways_columns  = ['type', 'thrower', 'throwerX', 'throwerY', 
                           'turnoverX', 'turnoverY', 'wasBlocked', 'team', 'throwawayID', 'year', 'gameID']
    drops_columns       = ['type', 'thrower', 'throwerX', 'throwerY', 
                           'receiver', 'receiverX', 'receiverY', 'team', 'throwawayID', 'year', 'gameID']
    callahans_columns   = ['type', 'thrower', 'throwerX', 'throwerY', 
                           'turnoverX', 'turnoverY', 'team', 'year', 'gameID'] # fix this
    stalls_columns      = ['type', 'thrower', 'throwerX', 'throwerY', 'team', 'year', 'gameID'] # see if this corresponds to blocks
    blocks_columns      = ['type', 'defender', 'blockX', 'blockY', 'team', 'blockID', 'year', 'gameID']

    # Helper to filter a dictionary by only keeping desired keys.
    def filter_dict(d, keys):
        return {k: d.get(k) for k in keys}
    
    # Define API endpoints.
    events_url = 'https://www.backend.ufastats.com/api/v1/gameEvents'
    games_url  = 'https://www.backend.ufastats.com/api/v1/games'
    
    # Get game details.
    game_ids_str = ','.join(game_ids)
    games_params = {'gameIDs': game_ids_str}
    try:
        games_response = requests.get(games_url, params=games_params)
        games_response.raise_for_status()
        games_data = games_response.json().get("data", [])
        team_id_map = {}
        for game in games_data:
            year = game["gameID"][:4]
            team_id_map[game["gameID"]] = {
                "home": game["homeTeamID"],
                "away": game["awayTeamID"],
                "year": year
            }
    except requests.exceptions.RequestException as e:
        print(f"Error fetching game data from {games_url}: {e}")
        return None

    # Final lists to hold data to be inserted.
    throws     = []  # use throws_columns
    goals      = []  # same as throws_columns
    throwaways = []  # use throwaways_columns
    drops      = []  # use drops_columns
    callahans  = []  # use callahans_columns
    stalls     = []  # use stalls_columns
    blocks     = []  # use blocks_columns

    # Process each game.
    for gameID in game_ids:
        params = {'gameID': gameID}
        try:
            response = requests.get(events_url, params=params)
            if response.status_code == 200 and response.headers.get('Content-Type', '').startswith('application/json'):
                data = response.json()
                # Get team IDs and year.
                home_team_id = team_id_map.get(gameID, {}).get("home", "Unknown")
                away_team_id = team_id_map.get(gameID, {}).get("away", "Unknown")
                game_year    = int(team_id_map.get(gameID, {}).get("year", "0"))
                
                home_events = data.get("data", {}).get("homeEvents", [])
                away_events = data.get("data", {}).get("awayEvents", [])
                
                # Tag each event with team, year, and gameID.
                for event in home_events:
                    event['team'] = home_team_id
                    event['year'] = game_year
                    event['gameID'] = gameID
                for event in away_events:
                    event['team'] = away_team_id
                    event['year'] = game_year
                    event['gameID'] = gameID
                
                events = home_events + away_events
                if not events:
                    continue
                # Create DataFrame to preserve ordering.
                df = pd.DataFrame(events)
                
                # Build dictionaries for turnovers and possession events for each team.
                # Turnovers: type 22 (throwaways) and type 20 (drops).
                # Possession events: type 13 (switch flags) and type 11 (blocks).
                turnovers_by_team = {home_team_id: [], away_team_id: []}
                poss_events_by_team = {home_team_id: [], away_team_id: []}
                turnover_counter = {home_team_id: 0, away_team_id: 0}  # assign throwawayID
                
                for i, row in df.iterrows():
                    team = row['team']
                    etype = row['type']
                    if etype in [22, 20]:
                        event_data = row.to_dict()
                        event_data['throwawayID'] = turnover_counter[team]
                        if etype == 22:
                            event_data['wasBlocked'] = 0
                        turnovers_by_team[team].append({
                            "orig_index": i,
                            "type": etype,
                            "turnoverID": turnover_counter[team],
                            "data": event_data
                        })
                        turnover_counter[team] += 1
                    elif etype in [13, 11]:
                        poss_event = {"orig_index": i, "type": etype, "data": row.to_dict()}
                        if etype == 11:
                            # For a block, look ahead to get additional info if needed.
                            next_row = df.iloc[i + 1]
                            if next_row['type'] == 3:
                                next_row = df.iloc[i+2]
                            poss_event["blockX"] = next_row['throwerX']
                            poss_event["blockY"] = next_row['throwerY']
                            poss_event["defender"] = row.get("defender")
                        poss_events_by_team[team].append(poss_event)
                    # it isnt tracked correctly when its an eoq event
                    # but I dont think we can just see if the previous event is a throwaway
                    # because it could be like that for both sides
                    # first thought would be that we need to backtrack to the previous pull, and the count the number of turns
#                    elif etype in [28, 29, 30, 31, 32, 33]:
                        # when we find an eoq event, we need to backtrack to the previous pull
                        # here, we build a dict of all throw and blocks
                        # with this dict, we can match find missing throws
                        
                        # actually, what we might have to do is go through our normal switch logic, then fill in
                        # dont think this can be done in line, because we need to look at the whole block?
#                        prev_row = df.iloc[i - 1]
#                        if prev_row['type'] ==  
                
                # Pair turnovers from one team with possession events from the opposing team.
                # Home team turnovers pair with away team possession events:
                home_turnovers = sorted(turnovers_by_team.get(home_team_id, []), key=lambda x: x["orig_index"])
                away_poss_events = sorted(poss_events_by_team.get(away_team_id, []), key=lambda x: x["orig_index"])
                for t_event, p_event in zip(home_turnovers, away_poss_events):
                    if p_event["type"] == 11:
                        t_event["data"]["wasBlocked"] = 1
                        # Use the turnover's coordinates for the block info:
                        t_event["data"]["blockX"] = t_event["data"].get("turnoverX")
                        t_event["data"]["blockY"] = t_event["data"].get("turnoverY")
                        t_event["data"]["blockDefender"] = p_event["defender"]
                        blocks.append({
                            "type": 11,
                            "defender": p_event["defender"],
                            "blockX": t_event["data"].get("turnoverX"),
                            "blockY": t_event["data"].get("turnoverY"),
                            "team": home_team_id,
                            "blockID": t_event["turnoverID"],
                            "year": t_event["data"]["year"],
                            "gameID": gameID
                        })
                    else:
                        t_event["data"]["wasBlocked"] = 0

                # Away team turnovers pair with home team possession events:
                away_turnovers = sorted(turnovers_by_team.get(away_team_id, []), key=lambda x: x["orig_index"])
                home_poss_events = sorted(poss_events_by_team.get(home_team_id, []), key=lambda x: x["orig_index"])
                for t_event, p_event in zip(away_turnovers, home_poss_events):
                    if p_event["type"] == 11:
                        t_event["data"]["wasBlocked"] = 1
                        t_event["data"]["blockX"] = t_event["data"].get("turnoverX")
                        t_event["data"]["blockY"] = t_event["data"].get("turnoverY")
                        t_event["data"]["blockDefender"] = p_event["defender"]
                        blocks.append({
                            "type": 11,
                            "defender": p_event["defender"],
                            "blockX": t_event["data"].get("turnoverX"),
                            "blockY": t_event["data"].get("turnoverY"),
                            "team": away_team_id,
                            "blockID": t_event["turnoverID"],
                            "year": t_event["data"]["year"],
                            "gameID": gameID
                        })
                    else:
                        t_event["data"]["wasBlocked"] = 0

                # Add turnovers to either throwaways or drops lists.
                for team in [home_team_id, away_team_id]:
                    for t_event in turnovers_by_team.get(team, []):
                        etype = t_event["type"]
                        event_data = t_event["data"]
                        if etype == 22:
                            throwaways.append(filter_dict(event_data, throwaways_columns))
                        elif etype == 20:
                            drops.append(filter_dict(event_data, drops_columns))
                
                # Process remaining events that are not turnovers or possession events.
                for i, row in df.iterrows():
                    etype = row['type']
                    team = row['team']
                    if etype == 18:
                        throws.append({
                            "type": etype,
                            "thrower": row.get("thrower"),
                            "throwerX": row.get("throwerX"),
                            "throwerY": row.get("throwerY"),
                            "receiver": row.get("receiver"),
                            "receiverX": row.get("receiverX"),
                            "receiverY": row.get("receiverY"),
                            "team": team,
                            "year": row.get("year"),
                            "gameID": gameID
                        })
                    elif etype == 19:
                        goals.append({
                            "type": etype,
                            "thrower": row.get("thrower"),
                            "throwerX": row.get("throwerX"),
                            "throwerY": row.get("throwerY"),
                            "receiver": row.get("receiver"),
                            "receiverX": row.get("receiverX"),
                            "receiverY": row.get("receiverY"),
                            "team": team,
                            "year": row.get("year"),
                            "gameID": gameID
                        })
                    elif etype == 23:
                        callahans.append({
                            "type": etype,
                            "thrower": row.get("thrower"),
                            "throwerX": row.get("throwerX"),
                            "throwerY": row.get("throwerY"),
                            "turnoverX": row.get("turnoverX"),
                            "turnoverY": row.get("turnoverY"),
                            "team": team,
                            "year": row.get("year"),
                            "gameID": gameID
                        })
                    elif etype == 24:
                        stalls.append({
                            "type": etype,
                            "thrower": row.get("thrower"),
                            "throwerX": row.get("throwerX"),
                            "throwerY": row.get("throwerY"),
                            "team": team,
                            "year": row.get("year"),
                            "gameID": gameID
                        })
        except requests.exceptions.RequestException as e:
            print(f"Error fetching event data from {events_url} for game {gameID}: {e}")
    
    return throws, goals, drops, throwaways, callahans, stalls, blocks

# need to account for eoq stats which dont show up
# I'm thinking going play by play makes this easier
```

```python
def check_mismatch(data):
    for gameID, events in data.items():
        print(f"Checking {gameID}:")
        _check_list(events)
        
def _check_list(events):
    for i, ev in enumerate(events):
        t = ev.get('type')
        if t == 22:
            if i+1 < len(events):
                nt = events[i+1].get('type')
                if nt not in (11, 13):
                    print(f"Mismatch at idx {i}: 22 → next is {nt!r}")
            else:
                print(f"Mismatch at idx {i}: 22 has no next event")
        if t in (11, 13):
            if i-1 >= 0:
                pt = events[i-1].get('type')
                if pt not in (20, 22):
                    print(f"Mismatch at idx {i}: {t} ← prev is {pt!r}")
            else:
                print(f"Mismatch at idx {i}: {t} has no previous event")
```

```python
FIELD_X_MIN, FIELD_X_MAX = -25.0, 25.0
FIELD_Y_MIN, FIELD_Y_MAX = 0.0, 110.0

def _clamp(v: float, lo: float, hi: float) -> float:
    return max(lo, min(hi, v))

def _jitter_endzone_xy(
    rng: random.Random,
    base_x: float = 0.0,
    base_y: float = 110.0,
    max_dx: float = 5.0,
    max_dy: float = 15.0,
) -> Tuple[float, float]:
    """
    Pick a turnover location near back-center of the end zone, with bounded jitter,
    then clamp to field bounds.
    """
    x = base_x + rng.uniform(-max_dx, max_dx)
    y = base_y + rng.uniform(-max_dy, max_dy)
    return float(_clamp(x, FIELD_X_MIN, FIELD_X_MAX)), float(_clamp(y, FIELD_Y_MIN, FIELD_Y_MAX))

def insert_missing_turnovers(
    combined_events: List[Dict],
    *,
    seed: Optional[int] = None
) -> List[Dict]:
    """
    Post-process a single game's combined event list and insert synthetic turnovers (type 22)
    when missing between a completed pass (type 18) and the next possession delimiter (type 28).

    Also enforces ordering so that defender credits (type 11) appear BEFORE type 28.

    Rule:
      - After a type 18, scan forward through any number of type 11 (defender credit) events.
      - If we encounter type 22 before type 28: do nothing.
      - If we encounter type 28 before type 22: insert a synthetic type 22 RIGHT BEFORE that 28.
      - Independently, ensure any '11' that would follow a '28' is moved before it.

    Returns
    -------
    List[Dict]
        New list with any needed type 22 events inserted and 11/28 order corrected.
    """
    rng = random.Random(seed)
    out: List[Dict] = []
    i = 0
    n = len(combined_events)

    while i < n:
        e = combined_events[i]
        out.append(e)

        # Ensure 11 always precedes 28 if they were adjacent in the source
        # (bubble the just-appended 11 leftwards across consecutive 28s).
        k = len(out) - 1
        while k > 0 and out[k].get("type") == 11 and out[k-1].get("type") == 28:
            out[k-1], out[k] = out[k], out[k-1]
            k -= 1

        if e.get("type") == 18:
            # Look ahead to find the next decisive event (22 or 28), allowing any number of 11s
            j = i + 1
            saw_22 = False
            insert_pos_in_out: Optional[int] = None

            while j < n:
                t = combined_events[j].get("type")
                if t == 22:
                    saw_22 = True
                    break
                if t == 28:
                    # Insert just before this 28 in the output stream (right after the 18 and any swaps done)
                    insert_pos_in_out = len(out)
                    break
                if t not in (11,):  # tolerate 11s; stop on other unexpected stuff
                    break
                j += 1

            if (not saw_22) and (insert_pos_in_out is not None):
                # Build synthetic 22 using the receiver of the 18
                thrower = e.get("receiver")
                if thrower is not None:
                    tX, tY = _jitter_endzone_xy(rng)
                    synth_22 = {
                        "type": 22,
                        "thrower": thrower,
                        "throwerX": e.get("receiverX"),
                        "throwerY": e.get("receiverY"),
                        "turnoverX": tX,
                        "turnoverY": tY,
                        "team": e.get("team"),
                        "year": e.get("year"),
                        "gameID": e.get("gameID"),
                    }
                    out.insert(insert_pos_in_out, synth_22)

        i += 1

    return out
```

```python
def fetch_and_write_players():
    """
    Fetches player data for years [2022, 2023, 2024] from the UFA Stats API,
    builds a DataFrame (all_players_df) with columns:
      [playerID, name, jerseyNumber, year, teamID],
    writes that DataFrame to all_players.csv,
    and returns the DataFrame.
    """
    years = [2022, 2023, 2024]
    csv_filename = "all_players.csv"
    base_url = "https://www.backend.ufastats.com/api/v1/players"

    # We'll collect all rows in a list of lists
    all_rows = []

    for year in years:
        try:
            response = requests.get(base_url, params={"years": year})
            response.raise_for_status()

            # The API response typically has {"object": "...", "data": [...]}
            data = response.json().get("data", [])
        except requests.exceptions.RequestException as e:
            print(f"Error fetching data for year={year}: {e}")
            continue

        # For each player, create rows for each team entry
        for player in data:
            player_id = player.get("playerID", "")
            first_name = player.get("firstName", "")
            last_name = player.get("lastName", "")
            full_name = f"{first_name} {last_name}".strip()

            # Each player can have multiple entries in "teams"
            for team_info in player.get("teams", []):
                team_id = team_info.get("teamID", "")
                team_year = team_info.get("year", "")
                jersey_number = team_info.get("jerseyNumber", "")
                
                if team_id == "allstars1" or team_id == "allstars2":
                    continue

                row = [
                    player_id,
                    full_name,
                    jersey_number,
                    team_year,
                    team_id
                ]
                all_rows.append(row)

    # Now build a DataFrame
    all_players_df = pd.DataFrame(
        all_rows,
        columns=["playerID", "name", "jerseyNumber", "year", "teamID"]
    )

    # Write the DataFrame to CSV
    all_players_df.to_csv(csv_filename, index=False)
    print(f"Done! Created '{csv_filename}' and built the DataFrame all_players_df.")

    return all_players_df
```

```python
def fetch_and_write_teams():
    """
    Fetches team data for years [2022, 2023, 2024] from the UFA Stats API,
    builds a DataFrame (all_teams_df) with columns:
      [teamID, fullName, year, divisionID, wins, losses, ties, standing],
    writes that DataFrame to all_teams.csv,
    and returns the DataFrame.

    Teams in the 'allstars' division are excluded.
    """
    years = [2022, 2023, 2024]
    csv_filename = "all_teams.csv"
    base_url = "https://www.backend.ufastats.com/api/v1/teams"

    all_rows = []

    for year in years:
        try:
            response = requests.get(base_url, params={"years": year})
            response.raise_for_status()

            # The API response shape: {"object": "list", "data": [...]}
            data = response.json().get("data", [])
        except requests.exceptions.RequestException as e:
            print(f"Error fetching data for year={year}: {e}")
            continue

        for team in data:
            # Extract fields
            team_id = team.get("teamID", "")
            full_name = team.get("fullName", "")
            team_year = team.get("year", "")
            division_id = team.get("division", {}).get("divisionID", "")
            wins = team.get("wins", "")
            losses = team.get("losses", "")
            ties = team.get("ties", "")
            standing = team.get("standing", "")

            # Skip teams in the 'allstars' division
            if division_id == "allstars":
                continue

            row = [
                team_id,
                full_name,
                team_year,
                division_id,
                wins,
                losses,
                ties,
                standing
            ]
            all_rows.append(row)

    # Build a DataFrame
    all_teams_df = pd.DataFrame(
        all_rows,
        columns=["teamID", "fullName", "year", "divisionID", "wins", "losses", "ties", "standing"]
    )

    # Write to CSV
    all_teams_df.to_csv(csv_filename, index=False)
    print(f"Done! Created '{csv_filename}' and built the DataFrame all_teams_df (excluding allstars).")

    return all_teams_df
```

```python
def create_and_insert_table(db_name, table_name, schema, data, columns):
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    
    # Drop the table if it exists
    cursor.execute(f'DROP TABLE IF EXISTS {table_name};')
    
    cursor.execute("PRAGMA foreign_keys = ON;")
    
    # Create the table with the provided schema
    cursor.execute(f'CREATE TABLE IF NOT EXISTS {table_name} ({schema});')
    conn.commit()
    
    # Prepare columns and placeholders for insertion
    column_names = ', '.join(columns)
    placeholders = ', '.join(['?'] * len(columns))
    insert_query = f'INSERT INTO {table_name} ({column_names}) VALUES ({placeholders})'
    
    # Prepare the data values
    data_list = []
    for row in data:
        # Ensure all required keys are present and in the correct order
        row_data = [row.get(column) for column in columns]
        data_list.append(tuple(row_data))
    
    # Insert the data into the table
    cursor.executemany(insert_query, data_list)
    conn.commit()
    
    # Close the connection
    cursor.close()
    conn.close()
```

```python
db_name = 'ufa_data.db'

# Years Table
years_table = 'years'
years_columns = ['year']
years_data = [{'year': 2022}, {'year': 2023}, {'year': 2024}]
years_table_schema = '''
    year INTEGER PRIMARY KEY
'''
create_and_insert_table(db_name, years_table, years_table_schema, years_data, years_columns)

# Teams Table with Composite Primary Key (teamID, year)
teams_table = 'teams'
teams_columns = ['teamID', 'fullName', 'year', 'divisionID', 'wins', 'losses', 'ties', 'standing']
teams_data = teams_data
teams_table_schema = '''
    teamID TEXT,
    fullName TEXT,
    year INTEGER,
    divisionID TEXT,
    wins INTEGER,
    losses INTEGER,
    ties INTEGER,
    standing INTEGER,
    PRIMARY KEY (teamID, year),
    FOREIGN KEY (year) REFERENCES years(year)
'''
create_and_insert_table(db_name, teams_table, teams_table_schema, teams_data, teams_columns)

# Players Table, now correctly referencing (teamID, year)
players_table = 'players'
players_columns = ['playerID', 'name', 'jerseyNumber', 'year', 'teamID']
players_data = players_data
players_table_schema = '''
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    playerID TEXT,
    name TEXT,
    jerseyNumber INTEGER,
    year INTEGER,
    teamID TEXT,
    FOREIGN KEY (teamID, year) REFERENCES teams(teamID, year)
'''
create_and_insert_table(db_name, players_table, players_table_schema, players_data, players_columns)
```

```python
db_name = 'ufa_data.db'

# Throws Table
throws_table = 'throws'
throws_columns = ['type', 'thrower', 'throwerX', 'throwerY', 'receiver', 'receiverX', 'receiverY', 'team', 'year', 'gameID']
throws_data = throws_data
throws_table_schema = '''
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    type INTEGER,
    thrower TEXT,
    throwerX REAL,
    throwerY REAL,
    receiver TEXT,
    receiverX REAL,
    receiverY REAL,
    team TEXT,
    year INTEGER,
    gameID TEXT,
    FOREIGN KEY (thrower, year) REFERENCES players (playerID, year),
    FOREIGN KEY (receiver, year) REFERENCES players (playerID, year),
    FOREIGN KEY (team, year) REFERENCES teams (teamID, year)
'''
create_and_insert_table(db_name, throws_table, throws_table_schema, throws_data, throws_columns)

# Goals Table (Fixed Schema)
goals_table = 'goals'
goals_columns = ['type', 'thrower', 'throwerX', 'throwerY', 'receiver', 'receiverX', 'receiverY', 'team', 'year', 'gameID']
goals_data = goals_data
goals_table_schema = '''
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    type INTEGER,
    thrower TEXT,
    throwerX REAL,
    throwerY REAL,
    receiver TEXT,
    receiverX REAL,
    receiverY REAL,
    team TEXT,
    year INTEGER,
    gameID TEXT,
    FOREIGN KEY (thrower, year) REFERENCES players (playerID, year),
    FOREIGN KEY (receiver, year) REFERENCES players (playerID, year),
    FOREIGN KEY (team, year) REFERENCES teams (teamID, year)
'''
create_and_insert_table(db_name, goals_table, goals_table_schema, goals_data, goals_columns)

# Throwaways Table
throwaways_table = 'throwaways'
throwaways_columns = ['type', 'thrower', 'throwerX', 'throwerY', 'turnoverX', 'turnoverY', 'wasBlocked', 'team', 'throwawayID', 'year', 'gameID']
throwaways_data = throwaways_data
throwaways_table_schema = '''
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    type INTEGER,
    thrower TEXT,
    throwerX REAL,
    throwerY REAL,
    turnoverX REAL,
    turnoverY REAL,
    wasBlocked INTEGER,
    team TEXT,
    throwawayID INTEGER,
    year INTEGER,
    gameID TEXT,
    FOREIGN KEY (thrower, year) REFERENCES players (playerID, year),
    FOREIGN KEY (team, year) REFERENCES teams (teamID, year)
'''
create_and_insert_table(db_name, throwaways_table, throwaways_table_schema, throwaways_data, throwaways_columns)

# Drops Table (Fixed Columns)
drops_table = 'drops'
drops_columns = ['type', 'thrower', 'throwerX', 'throwerY', 'receiver', 'receiverX', 'receiverY', 'team', 'throwawayID', 'year', 'gameID']
drops_data = drops_data
drops_table_schema = '''
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    type INTEGER,
    thrower TEXT,
    throwerX REAL,
    throwerY REAL,
    receiver TEXT,
    receiverX REAL,
    receiverY REAL,
    team TEXT,
    throwawayID INTEGER,
    year INTEGER,
    gameID TEXT,
    FOREIGN KEY (thrower, year) REFERENCES players (playerID, year),
    FOREIGN KEY (receiver, year) REFERENCES players (playerID, year),
    FOREIGN KEY (team, year) REFERENCES teams (teamID, year),
    FOREIGN KEY (throwawayID) REFERENCES throwaways (throwawayID)
'''
create_and_insert_table(db_name, drops_table, drops_table_schema, drops_data, drops_columns)

# Callahans Table
callahans_table = 'callahans'
callahans_columns = ['type', 'thrower', 'throwerX', 'throwerY', 'turnoverX', 'turnoverY', 'team', 'year', 'gameID']
callahans_data = callahans_data
callahans_table_schema = '''
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    type INTEGER,
    thrower TEXT,
    throwerX REAL,
    throwerY REAL,
    turnoverX REAL,
    turnoverY REAL,
    team TEXT,
    year INTEGER,
    gameID TEXT,
    FOREIGN KEY (thrower, year) REFERENCES players (playerID, year),
    FOREIGN KEY (team, year) REFERENCES teams (teamID, year)
'''
create_and_insert_table(db_name, callahans_table, callahans_table_schema, callahans_data, callahans_columns)

# Stalls Table
stalls_table = 'stalls'
stalls_columns = ['type', 'thrower', 'throwerX', 'throwerY', 'team', 'year', 'gameID']
stalls_data = stalls_data
stalls_table_schema = '''
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    type INTEGER,
    thrower TEXT,
    throwerX REAL,
    throwerY REAL,
    team TEXT,
    year INTEGER,
    gameID TEXT,
    FOREIGN KEY (thrower, year) REFERENCES players (playerID, year),
    FOREIGN KEY (team, year) REFERENCES teams (teamID, year)
'''
create_and_insert_table(db_name, stalls_table, stalls_table_schema, stalls_data, stalls_columns)

# Blocks Table
blocks_table = 'blocks'
blocks_columns = ['type', 'defender', 'blockX', 'blockY', 'team', 'blockID', 'year', 'gameID']
blocks_data = blocks_data
blocks_table_schema = '''
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    type INTEGER,
    defender TEXT,
    blockX REAL,
    blockY REAL,
    team TEXT,
    blockID INTEGER,
    year INTEGER,
    gameID TEXT,
    FOREIGN KEY (defender, year) REFERENCES players (playerID, year),
    FOREIGN KEY (blockID) REFERENCES throwaways (throwawayID),
    FOREIGN KEY (team, year) REFERENCES teams (teamID, year)
'''
create_and_insert_table(db_name, blocks_table, blocks_table_schema, blocks_data, blocks_columns)
```

```python
def execute_query_to_dataframe(db_path, query, params=None):
    """
    Executes an SQL query and returns the results as a pandas DataFrame.

    Args:
        db_path (str): Path to the SQLite database file.
        query (str): The SQL query to execute.
        params (tuple or dict, optional): Parameters to bind to the query.

    Returns:
        pd.DataFrame: DataFrame containing the query results.
    """
    # Connect to the database
    conn = sqlite3.connect(db_path)

    try:
        # Use pandas read_sql_query function
        if params:
            df = pd.read_sql_query(query, conn, params=params)
        else:
            df = pd.read_sql_query(query, conn)
        return df

    except Exception as e:
        print(f"An error occurred: {e}")
        return pd.DataFrame()

    finally:
        conn.close()
```

```python
def execute_non_query(db_path, query, params=None):
    """
    Executes a non-SELECT SQL query (e.g., INSERT, UPDATE, DELETE).

    Args:
        db_path (str): Path to the SQLite database file.
        query (str): The SQL query to execute.
        params (tuple or dict, optional): Parameters to bind to the query.

    Returns:
        int: Number of rows affected.
    """
    # Connect to the database
    conn = sqlite3.connect(db_path)

    try:
        cursor = conn.cursor()
        # Execute the query with parameters if provided
        if params:
            cursor.execute(query, params)
        else:
            cursor.execute(query)
        # Commit changes
        conn.commit()
        return cursor.rowcount  # Return the number of rows affected

    except Exception as e:
        print(f"An error occurred: {e}")
        return 0

    finally:
        conn.close()
```

## Hotspots

```python
def throwing_hotspots(df):
    # Ensure the DataFrame does not contain NaN values in the coordinate columns
    df_clean = df.dropna(subset=['throwerX', 'throwerY', 'receiverX', 'receiverY']).copy()
    
    # Plot hotspots for thrower positions
    plt.figure(figsize=(6, 10))  # Taller than wide
    sns.kdeplot(
        x=df_clean['throwerX'], y=df_clean['throwerY'], 
        shade=True, cmap='rainbow', 
        bw_adjust=0.5, levels=50
    )
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Hotspots for Thrower Positions')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.axhline(0, color='gray', lw=0.5, ls='--')
    plt.axvline(0, color='gray', lw=0.5, ls='--')
    plt.grid(True)
    plt.show()

    # Plot hotspots for receiver positions
    plt.figure(figsize=(6, 10))  # Taller than wide
    sns.kdeplot(
        x=df_clean['receiverX'], y=df_clean['receiverY'], 
        shade=True, cmap='rainbow', 
        bw_adjust=0.5, levels=50
    )
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Hotspots for Receiver Positions')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.axhline(0, color='gray', lw=0.5, ls='--')
    plt.axvline(0, color='gray', lw=0.5, ls='--')
    plt.grid(True)
    plt.show()
```

```python
throwing_hotspots(tbirds_throws)
```

```python
throwing_hotspots(execute_query_to_dataframe('ufa_db.db', 'SELECT * FROM goals WHERE team = ?', ('windchill',)))
```

```python
def block_hotspots(df):
    # Ensure the DataFrame does not contain NaN values in the block coordinate columns
    df_clean = df.dropna(subset=['blockX', 'blockY']).copy()
    
    # Plot hotspots for block positions
    plt.figure(figsize=(6, 10))  # Taller than wide
    
    # KDE plot
    sns.kdeplot(
        x=df_clean['blockX'], y=df_clean['blockY'], 
        shade=True, cmap='rainbow', 
        bw_adjust=0.5, levels=50
    )
    
    # Scatter plot of block positions
    plt.scatter(
        df_clean['blockX'], df_clean['blockY'], 
        color='black', alpha=0.7, s=10, label='Block Positions'
    )
    
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Hotspots for Block Positions')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.axhline(0, color='gray', lw=0.5, ls='--')
    plt.axvline(0, color='gray', lw=0.5, ls='--')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
block_hotspots(execute_query_to_dataframe('ufa_db.db', 'SELECT * FROM blocks WHERE team = ?', ('thunderbirds', )))
```

```python
def blocked_throws_hotspots(df):
    # Filter for throwaways that were blocked
    df_clean_turnovers = df[df['wasBlocked'] == 1].dropna(subset=['throwerX', 'throwerY']).copy()

    # Plot: Hotspots for turnover positions (wasBlocked = 1)
    plt.figure(figsize=(6, 10))
    sns.kdeplot(
        x=df_clean_turnovers['throwerX'], y=df_clean_turnovers['throwerY'], 
        shade=True, cmap='rainbow', 
        bw_adjust=0.5, levels=50
    )
    plt.scatter(
        df_clean_turnovers['throwerX'], df_clean_turnovers['throwerY'], 
        color='black', alpha=0.7, s=10, label='Turnover Positions (Blocked)'
    )
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Hotspots for Turnover Positions (Blocked)')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.axhline(0, color='gray', lw=0.5, ls='--')
    plt.axvline(0, color='gray', lw=0.5, ls='--')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
blocked_throws_hotspots(execute_query_to_dataframe('ufa_db.db', 'SELECT * FROM throwaways WHERE team = ?', ('thunderbirds', )))
```

```python
import plotly.express as px

def block_hotspots_interactive(df):
    # Ensure the DataFrame does not contain NaN values in the block coordinate columns
    df_clean = df.dropna(subset=['blockX', 'blockY']).copy()

    # Create density contour plot
    fig = px.density_contour(
        df_clean,
        x='blockX',
        y='blockY',
        title='Hotspots for Block Positions',
    )
    
    # Add scatter points with hover data
    fig.add_scatter(
        x=df_clean['blockX'],
        y=df_clean['blockY'],
        mode='markers',
        marker=dict(size=8, color='black', opacity=0.7),
        text=df_clean['defender'],  # Column with defender names
        hovertemplate="<b>Defender:</b> %{text}<br><b>X:</b> %{x}<br><b>Y:</b> %{y}<extra></extra>"
    )
    
    # Layout adjustments
    fig.update_layout(
        xaxis_title='X Coordinate',
        yaxis_title='Y Coordinate',
        xaxis=dict(showgrid=True, zeroline=True),
        yaxis=dict(showgrid=True, zeroline=True),
        height=600,
        width=400
    )
    
    fig.show()
```

```python
block_hotspots_interactive(execute_query_to_dataframe('ufa_db.db', 'SELECT * FROM blocks WHERE team = ?', ('thunderbirds', )))
```
