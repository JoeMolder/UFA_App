```python
def find_game_info(url):
    driver = webdriver.Chrome(service=service, options=chrome_options)
    
    # Open the URL
    driver.get(url)

    # Find and extract the team names
    team_elements = driver.find_elements(By.CLASS_NAME, "svelte-game-center-team-name")
    score_elements = driver.find_elements(By.CLASS_NAME, "svelte-game-center-score")
    
    game_data = driver.find_element(By.ID, "web-panel-game-center-stats")
    completion_data = game_data.find_elements(By.XPATH, '//td[text()="Completions"]/following-sibling::td')
    hucking_data = game_data.find_elements(By.XPATH, '//td[text()="Hucks"]/following-sibling::td')
    holding_data = game_data.find_elements(By.XPATH, '//td[text()="Hold %"]/following-sibling::td')
    
    team_comp_data = [completion_data[2].text, completion_data[3].text]
    team_hucking_data = [hucking_data[0].text, hucking_data[1].text]
    team_hold_data = [holding_data[0].text, holding_data[1].text]
    
    comp_percent = []
    completes = []
    attempts = []
    
    huck_percent = []
    huck_completes = []
    huck_attempts = []
    
    hold_percent = []
    hold_success = []
    hold_attempts = []
    
    for comp_data in team_comp_data:
        comp_split = [part.strip('()%') for part in comp_data.split()]
    
        percentage = int(comp_split[0])
        completions = int(comp_split[1].split('/')[0])
        attempted = int(comp_split[1].split('/')[1])
    
        comp_percent.append(percentage)
        completes.append(completions)
        attempts.append(attempted)
        
    for huck_data in team_hucking_data:
        huck_split = [part.strip('()%') for part in huck_data.split()]
    
        percentage = int(huck_split[0])
        completions = int(huck_split[1].split('/')[0])
        attempted = int(huck_split[1].split('/')[1])
    
        huck_percent.append(percentage)
        huck_completes.append(completions)
        huck_attempts.append(attempted)
        
    for hold_data in team_hold_data:
        hold_split = [part.strip('()%') for part in hold_data.split()]
    
        percentage = int(hold_split[0])
        completions = int(hold_split[1].split('/')[0])
        attempted = int(hold_split[1].split('/')[1])
    
        hold_percent.append(percentage)
        hold_success.append(completions)
        hold_attempts.append(attempted)
        
    team_names = [element.text for element in team_elements]
    scores = [element.text for element in score_elements]
    
    driver.quit()

    # Print the extracted team names
    #print("Name, Score, Throw Percentage, Completions, Attempts, Huck Percentage, Huck Completions, Huck Attempts")
    for team_name, score, comp_perc, complete, attempt, huck_perc, huck_complete, huck_attempt, hold_perc, hold_complete, hold_attempt in zip(team_names, scores, comp_percent, completes, attempts, huck_percent, huck_completes, huck_attempts, hold_percent, hold_success, hold_attempts):
        print('Name:', team_name, 'Score:', score, 'Throwing Data:', comp_perc, complete, attempt, 'Huck Data:', huck_perc, huck_complete, huck_attempt, 'Hold Data:', hold_perc, hold_complete, hold_attempt)
```

```python
def find_throw_data(url):
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.get(url)
    
    wait = WebDriverWait(driver, 5)  # Increased wait time
    play_by_play_container = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "play-by-play")))

    # Find all quarters
    quarters = play_by_play_container.find_elements(By.CLASS_NAME, "play-by-play-quarter")

    data = []

    for quarter in quarters:
        quarter_header = quarter.find_element(By.CLASS_NAME, "play-by-play-quarter-header").text
    
        # Find all play-by-play points
        play_points = quarter.find_elements(By.CLASS_NAME, "title")
        content_points = quarter.find_elements(By.CLASS_NAME, "content")
    
        for point, content in zip(play_points,content_points):
            play_by_play_score = point.find_element(By.CLASS_NAME, "play-by-play-score").text
            play_by_play_is_o_line = point.find_element(By.CLASS_NAME, "play-by-play-is-o-line").text
            play_by_play_clock = point.find_element(By.CLASS_NAME, "play-by-play-clock").text
            play_by_play_length = point.find_element(By.CLASS_NAME, "play-by-play-length").text
            play_by_play_roster = point.find_element(By.CLASS_NAME, "play-by-play-roster").text
            throws = []
            
            events = content.find_elements(By.CLASS_NAME, "play-by-play-event")
            for event in events:
                event_distance = event.find_element(By.CLASS_NAME, "play-by-play-event-distance").get_attribute("innerText").strip()
                event_label = event.find_element(By.CLASS_NAME, "play-by-play-event-label").get_attribute("innerText").strip()
                throws.append((event_distance, event_label))
                
            data.append({
                "quarter_header": quarter_header,
                "play_by_play_score": play_by_play_score,
                "play_by_play_is_o_line": play_by_play_is_o_line,
                "play_by_play_clock": play_by_play_clock,
                "play_by_play_length": play_by_play_length,
                "play_by_play_roster": play_by_play_roster,
                "event_details": throws
            })

    # Print the gathered data
    for entry in data:
        print(entry)
        print('\n')
        
    print(len(data))

    # Close the driver
    driver.quit()
```

```python
def find_throw_on_field(url): 
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.get(url)
    
    wait = WebDriverWait(driver, 5)
    
    allQuarters = driver.find_element(By.XPATH, '//*[@id="stats-pages"]/div[1]/div[5]/div[3]/div[1]/div/div[2]/div[3]/div[2]/button')
    allQuarters.click()
    
    trash = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "col-md-6")))
    field = trash.find_element(By.XPATH, "following-sibling::*")
    throws_on_field = field.find_elements(By.TAG_NAME, 'g')
    
    actions = ActionChains(driver)
    
    # get button position for quarters
    # run this twice, and have two separate tables
    
    throw_info = []
    catch_names = []
    throw_names = []
    
    # okay, so maybe we can select event and then run each seciton
        # 1. Throws and catches
        # 2. Goals and assists
        # 3. Throws and drops
        # 4. we are also gonna need a special case for callahans
            # I'm thinking maybe we can just add a goal to the other person stats as callahan and goal stat
            
    # alternatively, we could do player by player!
        # this way we could also have player throwing stats which would be sweet
    
    for g in throws_on_field:
        throw_positions = {}
        lines = g.find_elements(By.TAG_NAME, 'line')
        catches = g.find_elements(By.TAG_NAME, 'circle')
        throws = g.find_elements(By.TAG_NAME, 'rect')
        if lines:
            line = lines[0]
            color = line.get_attribute('stroke')
            x1 = round(float(line.get_attribute('x1')), 4)
            y1 = round(float(line.get_attribute('y1')), 4)
            x2 = round(float(line.get_attribute('x2')), 4)
            y2 = round(float(line.get_attribute('y2')), 4)
            dist = round(math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2) / 4, 2)
            if round(float(x1)) != 217 and not color == "black":
                throw_info.append({'color': color, 'x1': x1, 'y1': y1, 'x2': x2, 'y2': y2, 'dist': dist})
        
        if catches:
            catch = catches[0]
            if catch.get_attribute('stroke') == 'red' or catch.get_attribute('stroke') == 'green':
                actions.move_to_element(catch).perform()
            
                try:
                    popup = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".ui.popup.visible")))
                    popup_text = popup.find_element(By.CLASS_NAME, "content").text
                    catch_names.append({'Name': popup_text})
                    print(f"Moved to catch at coordinates: x = {round(float(catch.get_attribute('cx')), 4)}, y = {round(float(catch.get_attribute('cy')), 4)}")
                    print(popup_text)
                except Exception as e:
                    True
                    
        # we also need to check to see if a red line has the same x, y and if it does we need to check it as well
        # this accounts for throws to drops
        if throws:
            throw = throws[0]
            if throw.get_attribute('stroke') == 'red' or throw.get_attribute('stroke') == 'green' or throw.get_attribute('stroke') == 'darkred' or throw.get_attribute('stroke') == 'purple' or (
                throw.get_attribute('stroke') == 'lightblue' and any(
                line.get_attribute('stroke') == 'red' and
                (round(float(throw.get_attribute('x')), 4) - 5 <= round(float(line.get_attribute('x1')), 4) <= round(float(throw.get_attribute('x')), 4) + 5) and
                (round(float(throw.get_attribute('y')), 4) - 5 <= round(float(line.get_attribute('y1')), 4) <= round(float(throw.get_attribute('y')), 4) + 5)
                for line in lines)):
                actions.move_to_element(throw).perform()

                try:
                    popup = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".ui.popup.visible")))
                    popup_text = popup.find_element(By.CLASS_NAME, "content").text
                    throw_names.append({'Name': popup_text})
                    print(f"Moved to throw at coordinates: x = {round(float(throw.get_attribute('x')), 4)}, y = {round(float(throw.get_attribute('y')), 4)}")
                    print(popup_text)
                except Exception as e:
                    True
    
    for i in throw_info:
        print(i)
        
    print("Catch Events:")
    for i in catch_names:
        print(i)
        
    print("Throw Events")
    for i in throw_names:
        print(i)
```

```python
def find_team_player_stats(driver, field):
    stats = []
    
    wait = WebDriverWait(driver, 10)
    allPlayers = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="stats-pages"]/div[1]/div[5]/div[3]/div[1]/div/div[1]/div')))
    checkboxes = allPlayers.find_elements(By.CLASS_NAME, 'field')

    deselectAll = allPlayers.find_element(By.CLASS_NAME, 'deselect-all')
    deselectAll.click()

    for checkbox in checkboxes[2:]:
        try:
            label = checkbox.find_element(By.TAG_NAME, 'label')

            checkbox_input = checkbox.find_element(By.TAG_NAME, 'input')

            if not checkbox_input.is_selected():
                try:
                    checkbox_input.click()
                except ElementClickInterceptedException:
                    driver.execute_script("arguments[0].click();", checkbox_input)

            playerName = label.text

            throwsOnField = field.find_elements(By.TAG_NAME, 'g')
            
            actionsList = []

            #filled rect = throw
                # hollow rect = catch
            # filled circle = catch
                # hollow circle = throw
            
            for g in throwsOnField:
                if float(g.find_elements(By.TAG_NAME, 'rect')[0].get_attribute('x')) < 214:
                    
                    action = None
                    
                    line = g.find_elements(By.TAG_NAME, 'line')[0]
                    throw = g.find_elements(By.TAG_NAME, 'rect')[0]
                    try:
                        catch = g.find_elements(By.TAG_NAME, 'circle')[0]
                    except Exception as noCatch:
                        if throwType == 'purple':
                            x1, x2 = round(float(throw.get_attribute('x')), 4)
                            y2, y1 = round(float(throw.get_attribute('y')), 4)
                            action = 'stall'
                    
                    lineType = line.get_attribute('stroke')
                    throwType = throw.get_attribute('fill')
                    catchType = catch.get_attribute('fill')
                    
                    x1 = round(float(line.get_attribute('x1')), 4)
                    y1 = round(float(line.get_attribute('y1')), 4)
                    x2 = round(float(line.get_attribute('x2')), 4)
                    y2 = round(float(line.get_attribute('y2')), 4)
                            
                    
                    if throwType in ['lightblue', 'red', 'green', 'darkred']:
                        tag = 'throw'
                            
                        if throwType == 'lightblue' and lineType == 'red':
                            action = 'throw to drop'
                        elif throwType == 'red':
                            action = 'throwaway'
                        elif throwType == 'green':
                            action = 'assist'
                        elif throwType == 'darkred':
                            action = 'callahan'
                        
                        # throw location
                            # origin = x1, y1
                            # final = x2, y2
                            # tag = throw
                            # action = throwaway / throw to drop / assist / stall / callahan
                        
                        # for a throw, the following can happen
                            # blue square, blue line = completion
                            # blue square, red line = throw to drop
                            # red square, red line = throw away
                            # green square, green line = assist
                            # dark red square = callahan
                            # purple square = stall
                        
                    elif catchType in ['lightblue', 'green', 'red']:
                        tag = 'catch'
                        if catchType == 'red':
                            action = 'drop'
                        elif catchType == 'green':
                            action = 'goal'
                        
                        # catch location
                            # origin = x1, y1
                            # dest = x2, y2
                            # tag = catch
                            # action = goal / drop
                        
                        # following posibilities
                            # blue circle, blue line = reception
                            # red circle, red line = drop
                            # green circle, green line = goal
                    if not lineType == 'black':
                        actionsList.append({
                            'type': tag,
                            'origin': (x1, y1),
                            'dest': (x2, y2),
                            'specact': action
                        }) 
                        
                        
            if checkbox_input.is_selected():
                try:
                    checkbox_input.click()
                except ElementClickInterceptedException:
                    driver.execute_script("arguments[0].click();", checkbox_input)

            stats.append({'Player': playerName, 'Actions': actionsList})

        except NoSuchElementException:
            print("Label not found for this checkbox.")
    return stats

# We also need to add the player's name and game info
    # this way when we have the centralized database we can group by these values

# next step
    # we need to match up throws w/ times based on the text play by play
    # this will also allow us to find blocks
    
    # for this, im thinking:
        # we have throws from who to who, and can estimate the distance of throws and match off that
        # we should string together consecutive throws, then can match an entire play off that
```

```python
def print_stats(stats):
    """Prints the stats in a structured format."""
    for s in stats:
        print(f"Player: {s['Player']}")
        for action in s['Actions']:
            if action['type'] == 'throw':
                print(f"  Throw - Origin: {action['origin']}, Destination: {action['dest']}, Action: {action['specact']}")
            elif action['type'] == 'catch':
                print(f"  Catch - Origin: {action['origin']}, Destination: {action['dest']}, Action: {action['specact']}")
        print("-" * 40)
```

```python
def export_to_excel_per_player(stats, file_name="player_stats.xlsx"):
    """Exports the collected stats to an Excel file, with each player having their own sheet."""
    
    # next steps:
        # we need to find the players full name
            # im thinking to do this, we can get each teams roster, and match names
            # dont know if it needs to be separated by team, but we can cross that bridge when we get to it.
            # should be easy to match up becuase it already has initials for same last name players
    
    players_data = {}

    for stat in stats:
        player_name = stat['Player']
        if player_name not in players_data:
            players_data[player_name] = []
        
        for action in stat['Actions']:
            players_data[player_name].append({
                'Type': action['type'],
                'Origin': action['origin'],
                'Destination': action['dest'],
                'Action': action['specact']
            })

    with pd.ExcelWriter(file_name, engine='openpyxl') as writer:
        for player, actions in players_data.items():
            df = pd.DataFrame(actions)
            
            df.to_excel(writer, sheet_name=player[:31], index=False)
            
    print(f"Data exported successfully to {file_name} with individual sheets for each player")
```

```python
def find_all_player_stats(url):
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.get(url)

    wait = WebDriverWait(driver, 10)
    button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="stats-pages"]/div[1]/div[5]/div[3]/div[1]/div/div[2]/div[3]/div[2]/button')))
    button.click()

    wait = WebDriverWait(driver, 10)
    trash = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "col-md-6")))
    field = trash.find_element(By.XPATH, "following-sibling::*")

    stats = find_team_player_stats(driver, field)

    wait = WebDriverWait(driver, 10)
    button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="stats-pages"]/div[1]/div[5]/div[2]/div[1]/div[2]/button[2]')))
    button.click()

    stats += find_team_player_stats(driver, field)
    
    driver.quit()
    
    return stats
```

```python
def find_all_player_stats_not_sep(url):
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.get(url)

    wait = WebDriverWait(driver, 10)
    button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="stats-pages"]/div[1]/div[5]/div[3]/div[1]/div/div[2]/div[3]/div[2]/button')))
    button.click()

    wait = WebDriverWait(driver, 10)
    trash = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "col-md-6")))
    field = trash.find_element(By.XPATH, "following-sibling::*")

    stats = find_team_player_stats_not_sep(driver, field)

    wait = WebDriverWait(driver, 10)
    button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="stats-pages"]/div[1]/div[5]/div[2]/div[1]/div[2]/button[2]')))
    button.click()

    stats += find_team_player_stats_not_sep(driver, field)
    
    gameInfo = url.split('/')[-1]
    
    for stat in stats:
        stat['game'] = gameInfo
    
    driver.quit()
    
    return stats
```

```python
def stitch_possessions(possessions):
    stitched_possessions = []
    current_possession = possessions[0]  # Start with the first possession

    for i in range(1, len(possessions)):
        # Get the last throw of the current possession
        last_throw_current = current_possession['possession throws'][-1]
        
        if not isinstance(last_throw_current, tuple):
            # If the last element of the possession is not a tuple, finalize this possession
            stitched_possessions.append(current_possession)
            current_possession = possessions[i]
            continue
        
        stitched = False
        # Loop through subsequent possessions to check if they should be stitched
        for j in range(i, len(possessions)):
            first_throw_next = possessions[j]['possession throws'][0]
            
            if isinstance(first_throw_next, tuple):
                # Check if X coordinates match
                if last_throw_current[2][0] == first_throw_next[1][0]:
                    # Stitch the possessions
                    current_possession['possession throws'].extend(possessions[j]['possession throws'])
                    stitched = True
                    break
        
        if not stitched:
            # If no stitch happened, finalize the current possession
            stitched_possessions.append(current_possession)
            current_possession = possessions[i]

    # Append the last possession
    stitched_possessions.append(current_possession)

    return stitched_possessions
```

```python
def throwChain(throwList):
    
    throwsOnly = [throw for throw in throwList if throw['type'] == 'throw']
    
    origins = set()
    dests = set()
    
    # Collect all origins and destinations
    for throw in throwsOnly:
        origins.add(throw['origin'])
        dests.add(throw['dest'])
    
    playStarters = origins - dests  # Origins that don't have matching destinations
    
    #print(playStarters)  # Debugging to check playStarters
    
    possessionThrows = []
    
    # Loop through each starter in playStarters
    for firstThrow in playStarters:
        currentThrows = []  # Renamed from throwList for clarity
        
        # Find the actual throw object that matches the play starter's origin
        starterThrow = next((throw for throw in throwsOnly if throw['origin'] == firstThrow), None)
        
        if starterThrow:
            currentThrows.append((starterThrow['name'], starterThrow['origin'], starterThrow['dest']))  # Add the first throw to the list
            #currentThrows.append((starterThrow['name']))
            discPos = starterThrow['dest']  # Update disc position
            
            lastFound = False
            
            # Continue finding subsequent throws in the chain
            while not lastFound:
                
                nextThrow = next((throw for throw in throwsOnly if throw['origin'] == discPos), None)
                
                if nextThrow:  # Check if this throw continues the chain
                    currentThrows.append((nextThrow['name'], nextThrow['origin'], nextThrow['dest']))
                    #currentThrows.append((nextThrow['name']))
                    discPos = nextThrow['dest']  # Update disc position
                    
                    # If there's a special action, end the chain
                    if nextThrow['specact'] is not None:
                        if nextThrow['specact'] == 'throwaway':
                            currentThrows.append('throwaway')
                        elif nextThrow['specact'] == 'throw to drop':
                            currentThrows.append('Dropped by: ') # find the person who dropped, catch destination at these coords
                        elif nextThrow['specact'] == 'assist':
                            currentThrows.append('Assist to: ') # find who scored
                        lastFound = True
                else:
                    lastFound = True
            
            possessionThrows.append({'possession throws': currentThrows})
    
    fixedPossessions = stitch_possessions(possessionThrows)
    
    return fixedPossessions
```

```python
chains = throwChain(game_stats_not_sep)

totalThrows = 0;
numPlays = len(chains)

for chain in chains:
    print(chain)
    totalThrows += len(chain['possession throws'])
    print('\n')

print('Avg Throws:' + str(totalThrows/numPlays))
```
